```python
import torch
import matplotlib.pyplot as plt

from utils import get_rgb
from dataset import Sen2_MTC
```

```python
root = '/Users/sinaraoufi/Downloads/CTGAN/CTGAN/Sen2_MTC/dataset'
batch_size = 4

dataset = Sen2_MTC(root)
dataloader = torch.utils.data.DataLoader(
    dataset, batch_size=batch_size, shuffle=False, num_workers=4)
```

```python
fig, axes = plt.subplots(batch_size, 4, figsize=(10, 10))

images, cloud_free_images, _ = next(iter(dataloader))

for i, (ax1, ax2, ax3, ax4) in enumerate(axes):
    cloudy0, cloudy1, cloudy2 = get_rgb(images[0][i]), get_rgb(images[1][i]), get_rgb(images[2][i])
    cloud_free = get_rgb(cloud_free_images[i])

    ax1.imshow(cloudy0)
    ax1.set_title("Cloudy 0")
    ax1.axis('off')

    ax2.imshow(cloudy1)
    ax2.set_title("Cloudy 1")
    ax2.axis('off')

    ax3.imshow(cloudy2)
    ax3.set_title("Cloudy 2")
    ax3.axis('off')

    ax4.imshow(cloud_free)
    ax4.set_title("Cloud-free")
    ax4.axis('off')

plt.tight_layout()
plt.show()
```
